#### Reference Count
- primary memory manage method in python
- For every obj, python manages a ref count which points to its refernece obj
- Also if ref==0 means obj deallocated

```python
import sys

a = []

#2 means: ref for 'a' and getrefcount
print(sys.getrefcount(a))


#2 means: ref for 'a', 'b' and getrefcount
b = a
print(sys.getrefcount(a))
print(sys.getrefcount(b))

del b
print(sys.getrefcount(a))
```

#### Garbage Collection

- handles reference cycles, i.e. when objs ref each other then it prevents ref count from reaching 0


---

```python
import gc

# Simple Functions
gc.enable()
# gc.disable()
print(gc.collect())

# get Garbage Collection Stats
print(gc.get_stats())

# get unreachable objects
print(gc.garbage)

```
