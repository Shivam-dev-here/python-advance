#### Reference Count
- primary memory manage method in python
- For every obj, python manages a ref count which points to its refernece obj
- Also if ref==0 means obj deallocated

```python
import sys

a = []

#2 means: ref for 'a' and getrefcount
print(sys.getrefcount(a))


#2 means: ref for 'a', 'b' and getrefcount
b = a
print(sys.getrefcount(a))
print(sys.getrefcount(b))

del b
print(sys.getrefcount(a))
```

#### Garbage Collection

- handles reference cycles, i.e. when objs ref each other then it prevents ref count from reaching 0

- Basic functions
    1. import gc
    2. gc.enable()
    3. gc.disable()
    4. gc.collect()

- get Garbage Collection Stats: `gc.get_stats()`
- get unreachable objects: `gc.garbage`

---

##### Example: Using a Cyclic Reference
- Problem: Cyclic, create c, create h, del c, create c, del h, create h, repeat
- Solution: Manual trigger gc

```python
import gc

class MyObject:
    def __init__(self, name):
        self.name = name
        print(f"{self.name} Created")

    def __del__(self):
        print(f"{self.name} Deleted")

college = MyObject("Hw/Assigenment")
hostel = MyObject("Code/Sleep")

college = hostel
hostel = college

del college
del hostel

# Manual Triggering GC

print(gc.collect())
print(gc.garbage)
```
