# Logging: Multiple Loggers

- We can create multiple loggers for every part or modules of a applications

---

```python
import logging

## Creating logging for module1
logger1 = logging.getLogger("module1")
logger1.setLevel(logging.DEBUG)

## Creating logging for module2
logger2 = logging.getLogger("module2")
logger2.setLevel(logging.DEBUG)

# Cofigure Logging Settings

logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)

logger1.debug("Successful")
logger2.info("System working normal")
logger2.critical("function is undefined")
```

---
- 2026-06-27 21:14:00 - module1 - DEBUG - Successful
- 2026-06-27 21:14:00 - module2 - INFO - System working normal
- 2026-06-27 21:14:00 - module2 - CRITICAL - function is undefined

here instead of root, we are receiving module1 and module2
