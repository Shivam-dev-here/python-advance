# Logging

- Provides a way to track events, errors and other operational information.

- Python have built in logging modules offers a flexible framework for emitting log messages from python programs

---

#### Logging Levels

- DEBUG: Detailed information, diagnossing a problem.
- INFO: Confirmation for working properly.
- WARNING: Something unexpected (has/will) happen like ('disk space log'), But software running properly.
- ERROR: Some serious problem caused no execution to some function.
- CRITICAL: Some serious problem may caused the program to stop.
