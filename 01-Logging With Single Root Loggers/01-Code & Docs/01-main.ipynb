# Logging

- Provides a way to track events, errors and other operational information.

- Python have built in logging modules offers a flexible framework for emitting log messages from python programs

---

#### Logging Levels

- DEBUG: Detailed information, diagnossing a problem.
- INFO: Confirmation for working properly.
- WARNING: Something unexpected (has/will) happen like ('disk space log'), But software running properly.
- ERROR: Some serious problem caused no execution to some function.
- CRITICAL: Some serious problem may caused the program to stop.

- Example: `logging.debug("DEBUG: Completed")`
Similary other

---

#### Importing & Configuring basic log setting

- Must restart kernel in case want to use different configure setting
---

```python
import logging

# Configure basic logging settings
logging.basicConfig(level=logging.DEBUG)

# Logging Level Messages
logging.debug(" This is a debug msg")

```

---
#### Example: Configure Logging Setting with Custom Format
- Correct `datefmt='%Y-%m-%d %H-%M-%S'`
- Incorrect `datefmt='%Y-%M-%D %H-%M-%S'`
---

```python
import logging

logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H-%M-%S'
)

logging.debug("This is a debug msg")
```

---
#### Writing a file via config setting
---

```python
import logging

logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H-%M-%S'
)

logging.debug("Successful")
logging.info("System working normal")
logging.warning("value expected a list")
logging.error("missing a ','")
logging.critical("function is undefined")
```
